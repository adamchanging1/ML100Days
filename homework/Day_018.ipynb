```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)

plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.plot(X, Y, "b^")
plt.show()
```
